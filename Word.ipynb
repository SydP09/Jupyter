```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Define the text containing the words you want to include in the word cloud
text = "Python Developer C++ C AWS GCP Amazon Remote Success Failure Automation GCP Word Excel 0.0022 VKV KV KU Poetry Messi Adesanya Sales Amazon Microsoft Script Google Money INFJ Coding Programming Data Science Music Spainish Language Moma Marketing Digital  Social Media Campus Internship Skills Reach Event Strategy Remote Computer Science Engineering  IYS WSDA Machine  Content Writing Reading Writing Team Player Leadership Football AWS Learning Coursera CS50 Algorithm AI Machine Learning Learner LinkedIN API SQl Git Json Pandas SciPy Machine Vision  Agile Ambassador Volunteer"

# Define the custom color function
def custom_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    colors = {
        'Python': 'red',
        'C++': 'green',
        'AWS': 'yellow',
        'GCP': 'blue',
        'Amazon': 'purple',
        'Microsoft': 'orange',
    }
    return colors.get(word, 'black')  # Return the color from the mapping, or black as default

# Create a WordCloud object with the custom color function
wordcloud = WordCloud(width=1584, height=396, background_color='white', color_func=custom_color_func).generate(text)

# Display the word cloud with the custom font colors
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud.recolor(colormap='viridis', random_state=42), interpolation='bilinear')
plt.axis('off')
plt.show()
```
